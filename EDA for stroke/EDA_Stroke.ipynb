```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
stroke_df = pd.read_csv('stroke_data.csv')
```

```python
print(stroke_df.shape)
stroke_df.head()
```

```python
stroke_df.info()
```

```python
corrs = stroke_df.corr()

plt.figure(figsize=(12, 12))
sns.heatmap(corrs, annot=True)
plt.show()
```

```python
tab_df = stroke_df.copy()
```

```python
stroke_df.head()
```

```python
tab_df["sex"] = tab_df["sex"].map({0: "Female", 1: "Male"})
tab_df["hypertension"] = tab_df["hypertension"].map({0: "No", 1: "Yes"})
tab_df["heart_disease"] = tab_df["heart_disease"].map({0: "No", 1: "Yes"})
tab_df["ever_married"] = tab_df["ever_married"].map({0: "No", 1: "Yes"})
tab_df["work_type"] = tab_df["work_type"].map({0: "Never_worked", 1: "children", 2: "Govt_job", 3: "Self-employed", 4: "Private"})
tab_df["Residence_type"] = tab_df["Residence_type"].map({0: "Urban", 1: "Rural"})
tab_df["smoking_status"] = tab_df["smoking_status"].map({0: "No", 1: "Yes"})
tab_df["stroke"] = tab_df["stroke"].map({0: "No", 1: "Yes"})

tab_df.head()
```

```python
tab_df.info()
```

```python
tab_df["age"] = tab_df[["age"]].astype(int)
```

```python
tab_df["age"].describe()
```

```python
tab_df = tab_df[tab_df["age"] >= 0]
```

```python
tab_df["age"].describe()
```

```python
tab_df.info()
```

```python
tab_df.dropna(inplace=True)
```

```python
tab_df.info()
```

```python
tab_df.to_csv('clean_stroke_data.csv')
```
